# Algorithm #1 = Brute Force (Generic)

```python
from itertools import permutations

def solve_cryptarithmetic():
    letters = {'S', 'E', 'N', 'D', 'M', 'O', 'R', 'Y'}
    for perm in permutations('0123456789', len(letters)):
        assignment = dict(zip(letters, map(int, perm)))

        if assignment['M'] != 1:
            continue

        send = assignment['S'] * 1000 + assignment['E'] * 100 + assignment['N'] * 10 + assignment['D']
        more = assignment['M'] * 1000 + assignment['O'] * 100 + assignment['R'] * 10 + assignment['E']
        money = assignment['M'] * 10000 + assignment['O'] * 1000 + assignment['N'] * 100 + assignment['E'] * 10 + assignment['Y']

        if send + more == money:
            print("Solution found:")
            print(f"   {assignment['S']} {assignment['E']} {assignment['N']} {assignment['D']}")
            print(f"+  {assignment['M']} {assignment['O']} {assignment['R']} {assignment['E']}")
            print("--------------")
            print(f"=  {assignment['M']} {assignment['O']} {assignment['N']} {assignment['E']} {assignment['Y']}")
            return

    print("No solution found.")

if __name__ == "__main__":
    solve_cryptarithmetic()
```

# Algorithm 2 = Brute Force (DFS)

```python
import time

def solve_cryptarithmetic():
    letters = ['S', 'E', 'N', 'D', 'M', 'O', 'R', 'Y']
    count = 0  # Counter to limit the number of steps displayed
    start_time = time.time()  # Record start time

    def dfs(assignment):
        nonlocal count
        if 'M' in assignment and assignment['M'] != 1:
            return

        if len(assignment) == len(letters):
            send = assignment['S'] * 1000 + assignment['E'] * 100 + assignment['N'] * 10 + assignment['D']
            more = assignment['M'] * 1000 + assignment['O'] * 100 + assignment['R'] * 10 + assignment['E']
            money = assignment['M'] * 10000 + assignment['O'] * 1000 + assignment['N'] * 100 + assignment['E'] * 10 + assignment['Y']

            if send + more == money:
                print("Solution found:")
                print(f"   {assignment['S']} {assignment['E']} {assignment['N']} {assignment['D']}")
                print(f"+  {assignment['M']} {assignment['O']} {assignment['R']} {assignment['E']}")
                print("--------------")
                print(f"=  {assignment['M']} {assignment['O']} {assignment['N']} {assignment['E']} {assignment['Y']}")
                end_time = time.time()  # Record end time
                print(f"Time taken: {end_time - start_time} seconds")
                exit(0)  # Terminate the program when a solution is found

        count += 1
        if count % 10000000 == 0:
            print(f"Step {count}: {assignment}")

        for letter in letters:
            if letter not in assignment:
                for digit in range(10):
                    if digit not in assignment.values():
                        new_assignment = assignment.copy()
                        new_assignment[letter] = digit
                        dfs(new_assignment)

    dfs({})

    end_time = time.time()  # Record end time
    print(f"Time taken: {end_time - start_time} seconds")
    print("No solution found.")

if __name__ == "__main__":
    solve_cryptarithmetic()
```

# Algorithm #3 (Heuristic)

```python
from itertools import permutations
from queue import PriorityQueue

def heuristic(state):
    unassigned = sum(1 for digit in state if digit is None)
    return unassigned

def solve_cryptarithmetic():
    letters = ('S', 'E', 'N', 'D', 'M', 'O', 'R', 'Y')
    start_state = [None] * len(letters)  
    goal_state = None
    
    frontier = PriorityQueue()
    frontier.put((heuristic(start_state), start_state))

    while not frontier.empty():
        _, current_state = frontier.get()
        
        for perm in permutations('0123456789', len(letters)):
            new_state = current_state.copy()
            new_assignments = dict(zip(letters, map(int, perm)))

            if new_assignments['M'] != 1:
                continue

            send = new_assignments['S'] * 1000 + new_assignments['E'] * 100 + new_assignments['N'] * 10 + new_assignments['D']
            more = new_assignments['M'] * 1000 + new_assignments['O'] * 100 + new_assignments['R'] * 10 + new_assignments['E']
            money = new_assignments['M'] * 10000 + new_assignments['O'] * 1000 + new_assignments['N'] * 100 + new_assignments['E'] * 10 + new_assignments['Y']

            if send + more == money:
                goal_state = new_assignments
                break

            if new_state.count(None) > 0:
                frontier.put((heuristic(new_state), new_state))

        if goal_state:
            break

    if goal_state:
        print("Solution found:")
        print(f"   {goal_state['S']} {goal_state['E']} {goal_state['N']} {goal_state['D']}")
        print(f"+  {goal_state['M']} {goal_state['O']} {goal_state['R']} {goal_state['E']}")
        print("--------------")
        print(f"=  {goal_state['M']} {goal_state['O']} {goal_state['N']} {goal_state['E']} {goal_state['Y']}")
    else:
        print("No solution found.")

if __name__ == "__main__":
    solve_cryptarithmetic()
```
